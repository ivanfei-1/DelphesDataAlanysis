```python
# import dependencies
import ROOT as root
import numpy as np
import matplotlib.pyplot as plt

# open the file
fileName = "tag_1_delphes_events.root"
directory = str("tag_1_delphes_events.root")
File = root.TChain("Delphes;1")
File.Add(directory)
Number = File.GetEntries()
```

```python

PT = []
for i in range(Number):
    Entry = File.GetEntry(i)
    EntryFromBranch = File.Jet.GetEntries()
    for j in range(EntryFromBranch):
        PT.append(File.GetLeaf("Jet.PT").GetValue(j))

# plot the list of PT
print(PT)
plt.hist(PT, bins=100)
```
